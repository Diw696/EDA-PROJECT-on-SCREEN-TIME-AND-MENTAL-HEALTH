creating new col for better understanding for the social time hr and also for the numeric valeus

```python
def clean_social_time(x):
    try:
        # Try converting to float directly
        return float(str(x).strip())
    except:
        x = str(x).lower().strip()
        if 'less' in x:
            return 0.5
        if 'more than 5' in x:
            return 6
        if 'between' in x:
            parts = x.replace('hours', '').replace('hour', '').split('and')
            if len(parts) == 2:
                try:
                    low = float(parts[0].split()[-1])
                    high = float(parts[1].split()[0])
                    return (low + high) / 2
                except:
                    return None
    return None
```

```python
df_social_media['social_time_hr'] = df_social_media['8. What is the average time you spend on social media every day?'].apply(clean_social_time)

print(df_social_media['social_time_hr'].unique())
print(df_social_media[['8. What is the average time you spend on social media every day?', 'social_time_hr']].head())
```
